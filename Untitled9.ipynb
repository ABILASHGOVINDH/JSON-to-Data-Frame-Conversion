```python
import json
import pandas as pd
from pathlib import Path


json_files = [
    r"C:\Users\avina\Downloads\json\file1.json",
    r"C:\Users\avina\Downloads\json\file2.json",
    r"C:\Users\avina\Downloads\json\file3.json",
    r"C:\Users\avina\Downloads\json\file4.json",
    r"C:\Users\avina\Downloads\json\file5.json"
]


def load_json(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return json.load(f)


data_list = [load_json(file) for file in json_files]


def extract_restaurants(data):
    all_restaurants = []
    for page in data:
        if "restaurants" in page and isinstance(page["restaurants"], list):
            all_restaurants.extend([restaurant["restaurant"] for restaurant in page["restaurants"] if isinstance(restaurant, dict) and "restaurant" in restaurant])
    return all_restaurants


restaurants_list = [extract_restaurants(data) for data in data_list]


all_restaurants = [restaurant for sublist in restaurants_list for restaurant in sublist]


def normalize_restaurants(restaurants):
    return pd.json_normalize(restaurants)


df = normalize_restaurants(all_restaurants)


csv_path = r"C:\Users\avina\Downloads\restaurants_data.csv"
df.to_csv(csv_path, index=False, encoding="utf-8")

print(f"CSV file saved successfully at: {csv_path}")
```
